*Loc e Iloc*
loc: permite pegar uma linha de acordo com indice dela. Ele dá erro caso nao encontre o indice. loc[indice_linha, indice_coluna]

iloc: exerga o dataframe como linhas e colunas, e se consegue acessar o calor atraves do numero do indice e da coluna, apenas a posicao importa. iloc[num_linha, num_coluna]

```python
import pandas as pd
novo_prdodutos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';')
display(novo_prdodutos_df.head())
```

```python
#metodo loc
print(novo_prdodutos_df.loc[0, 'Custo Unitario'])
#metodo iloc 
print(novo_prdodutos_df.iloc[0, 2])
```

Se eu quisesse alterar o preco do produto com Id 873, como faria?

```python
novo_prdodutos_df.loc[0, 'Preco Unitario'] = 25
display(novo_prdodutos_df)
```
