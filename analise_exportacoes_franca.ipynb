### Importando base de dados


```python
import pandas as pd

exportacoes_franca_df = pd.read_csv('exportacoes_franca.csv')
exportacoes_franca_df = exportacoes_franca_df.loc[exportacoes_franca_df['Economic Block'] == 'Europe', :]
display(exportacoes_franca_df)
```

### Informações gerais


```python
print(exportacoes_franca_df.info())
#print(exportacoes_franca_df.describe())
```

### Como foi a evolução das exportações para França ao longo dos anos?


```python
tabela_export_anos = exportacoes_franca_df.groupby("Year").sum()
tabela_export_anos = tabela_export_anos[['US$ FOB']]

#criar um funcao de formatacao

def formatar(valor):
    return f"${valor:,.2f}"

#aplica formatacao

tabela_export_anos['US$ FOB'] = tabela_export_anos['US$ FOB'].map(formatar)
display(tabela_export_anos)
```

#### Quais produtos foram mais exportados em (US$) ao longo do período?

```python
export_por_produto = exportacoes_franca_df.groupby('SH4 Description').sum()
export_por_produto = export_por_produto[['US$ FOB']]

#coloca os valores em ordem decrescente
export_por_produto = export_por_produto.sort_values('US$ FOB', ascending= False)

#aplica formatacao
export_por_produto['US$ FOB'] = export_por_produto['US$ FOB'].map(formatar)
display(export_por_produto)
```

### Em 2020, qual cidade exportou mais para a França?

```python
tabela_2020 = exportacoes_franca_df.loc[exportacoes_franca_df["Year"] == 2020, :]
#display(tabela_2020)

export_por_cidade = tabela_2020.groupby('City').sum()
export_por_cidade = export_por_cidade[['US$ FOB']]
export_por_cidade = export_por_cidade.sort_values('US$ FOB', ascending= False)

export_por_cidade['US$ FOB'] = export_por_cidade['US$ FOB'].map(formatar)
display(export_por_cidade)

```

### O que a maior a maior cidade exportadora exporta pra França?

```python
tabela_cidade_top1 = tabela_2020.loc[tabela_2020["City"] == 'Duque de Caxias - RJ', :]
#display(tabela_2020)

tabela_cidade_top1 = tabela_2020.groupby('SH4 Description').sum()
tabela_cidade_top1 = tabela_cidade_top1[['US$ FOB']]
tabela_cidade_top1 = tabela_cidade_top1.sort_values('US$ FOB', ascending= False)

tabela_cidade_top1['US$ FOB'] = tabela_cidade_top1['US$ FOB'].map(formatar)
display(tabela_cidade_top1)
```
