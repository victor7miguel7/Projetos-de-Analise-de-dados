Preparando base de dados


```python
import pandas as pd
#importando arquivos
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv('Contoso - Lojas.csv', sep=';')
clientes_df = pd.read_csv('Contoso - Clientes.csv', sep=';')

#limpando apenas para colunas que queremos

clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Categoria']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

#mesclando e renomeando dataframes

vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on= 'ID Cliente').rename(columns={'E-mail':'E-mail Cliente'})
display(vendas_df)
```

Qual o % das vendas que foi devolvido?(Total devolvido/ Total Vendido
)

```python
qnt_vendida = vendas_df['Quantidade Vendida'].sum()
qnt_devolvida = vendas_df['Quantidade Devolvida'].sum()

print('{:.2%}'.format(qnt_devolvida / qnt_vendida))
```

Agora, se quisermos fazer a mesma análise apenas para 1 loja. Queremos filtrar apenas os itens da Loja Europe Online e saber % de devolução dessa loja.

```python
so_loja_europe = vendas_df[vendas_df['ID Loja'] == 306]
#display(so_loja_europe)
qnt_vendida_europe = so_loja_europe['Quantidade Vendida'].sum()
qnt_devolvida_europe = so_loja_europe['Quantidade Devolvida'].sum()

print('{:.2%}'.format(qnt_devolvida_europe/qnt_vendida_europe))
```

Se eu quisesse criar uma tabela apenas com as vendas Contoso Europe Online e que nao teveram nenhuma devolução. Quero criar essa tabela e saber quantas vendas são.


```python
so_loja_europe_sem_devolucao = vendas_df[(vendas_df['ID Loja']==306) & (vendas_df['Quantidade Devolvida']==0)]
display(so_loja_europe_sem_devolucao)
```
